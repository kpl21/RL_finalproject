Small Grid:


20x20 cell grid\
States: each cell corresponds to a single state, each state is a coordinate on the grid, (y,x). 

Actions: keyboard movement [RIGHT,LEFT,UP,DOWN] which corresponds to [0,1,2,3]

Goal: green cell located at (19,19), agent receives +100 reward when reaching the goal

Challenges:\
    Lava: orange cells, agent receives -5 reward upon landing\
    Hole: dark gray cells, agent receives -10 reward and then gets sent back to the starting point\
    Wall: light gray cells, agent cannot move to these walls\
    Stochastic Wind: light green cells, agent is moved an additional [-1,0,1] which is randomly chosen + its current action

Rewards:\
    Lava: -5\
    Hole: -10\
    Goal: +100
    Every other valid state: -0.25
    
Desired Behavior:\
Agent should reach the goal while avoiding all the obstacles within a fewest amount of steps after learning

Since this is a gridworld, all the states are discrete thus any tabular method of tracking the Q are the best. In this case we use TD such as: SARSA, Q-Learning, Double Q-Learning. We would not want to use Monte Carlo since it would require reaching the end before we update which would require us to loop through the entire state history again which would require additional running time. We chose to use Q-Learning since this is an off policy algorithm and although Sarsa may be better in terms of maximizing the rewards it takes longer to learn. We originally were using Double Q-Learning however we realized that since this required keeping track of 2 Qs that it takes longer to converge. We found Q-Learning to be the best tabular learning algorithm for our gridworlds.
