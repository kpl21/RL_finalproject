Small Grid


15x15 cell grid\
States: each cell corresponds to a single state, each state is a coordinate on the grid, (y,x). 

Actions: keyboard movement [RIGHT,LEFT,UP,DOWN] which corresponds to [0,1,2,3]

Goal: green cell located at (0,12), agent receives +0 reward when reaching the goal

Challenges:\
    Lava: red cells, agent receives -5 reward upon landing\
    Wall: light gray cells, agent cannot move to these walls\

Rewards:\
    Lava: -5\
    Goal: +0
    Every other valid state: -0.1
    
Desired Behavior:

Agent should reach the goal while avoiding all the obstacles within a fewest amount of steps after learning

Large Grid
